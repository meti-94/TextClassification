```
tag_scheme = [
            'B-Country',
            'B-Date',
            'B-DateAndTime',
            'B-Event',
            'B-Facility',
            'B-GeoPoliticalLocation',
            'B-Location',
            'B-NORP',
            'B-Organization',
            'B-Person',
            'I-Country',
            'I-Date',
            'I-DateAndTime',
            'I-Event',
            'I-Facility',
            'I-GeoPoliticalLocation',
            'I-Location',
            'I-NORP',
            'I-Organization',
            'I-Person'
 ]
all_ligal_tags = set(tag_scheme+['O'])
def read_data(path):
  sentences, tags = [], []
  with open(path, 'r', encoding='utf-8') as fin:
    sent_temp, tag_temp = [], []
    for line in fin:
      line = line.strip()
      line = line.replace('B-Date', 'B-DateAndTime')
      line = line.replace('I-Date', 'I-DateAndTime')
      if line=='':
        if (len(sent_temp)>=3 and len(tag_temp)>=3) and (set(tag_temp).issubset(all_ligal_tags)):
          sentences.append(sent_temp)
          tags.append(tag_temp)
        sent_temp, tag_temp = [], []
      else:
        words = line.split()
        words = list(map(lambda string:string.strip(), words))
        if len(words)==2:
          sent_temp.append(words[0].strip()); tag_temp.append(words[1])
        if len(words)==3:
          sent_temp.append(' '.join(words[:2])); tag_temp.append(words[2])
  return {'sentences':sentences, 'tags':tags}

def get_unique_tags(path):
  tags = []
  with open(path, 'r', encoding='utf-8') as fin:
    for line in fin:
      line = line.strip()
      if line=='':
        continue
      else:
        words = line.split()
        if len(words)==2:
          tags.append(words[1])
        if len(words)==3:
          tags.append(words[2])
  return set(tags)
```

```
training = read_data('/content/drive/MyDrive/Labeled_data_train.txt')
validation = read_data('/content/drive/MyDrive/Labeled_data_dev.txt')
```

```
tag_scheme = [
            'B-Country',
            # 'B-Date',
            'B-DateAndTime',
            'B-Event',
            'B-Facility',
            'B-GeoPoliticalLocation',
            'B-Location',
            'B-NORP',
            'B-Organization',
            'B-Person',
            'I-Country',
            # 'I-Date',
            'I-DateAndTime',
            'I-Event',
            'I-Facility',
            'I-GeoPoliticalLocation',
            'I-Location',
            'I-NORP',
            'I-Organization',
            'I-Person'
 ]
```

```
transformer =  'HooshvareLab/bert-fa-base-uncased' #'bert-base-multilingual-uncased'
```

```
# hyperparameters for network
dropout = 0.1
# hyperparameters for training
training_hyperparameters = {'epochs' : 6,'warmup_steps' : 500,'train_batch_size':32 ,'learning_rate': 0.00002}
```

```
from NERDA.models import NERDA
model = NERDA( dataset_training = training,
              dataset_validation = validation,
              tag_scheme = tag_scheme, 
              tag_outside = 'O',
              transformer = transformer,
              dropout = dropout,
              hyperparameters = training_hyperparameters,
              max_len=150)
```

```
model.train()
```

```
model.evaluate_performance(validation)
```
